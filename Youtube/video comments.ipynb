```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import time 
import pandas as pd
```

# ÏïÑÏù¥Îîî, ÎåìÍ∏Ä
## css selector

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.youtube.com/watch?v=_II0fc7hgNY')

time.sleep(3) 

# Ïä§ÌÅ¨Î°§ ÎÇ¥Î¶¨Í∏∞
e = driver.find_element_by_tag_name('body')
for i in range(30): # 30Î≤à
    e.send_keys(Keys.PAGE_DOWN)
    time.sleep(2) 
```

```python
users = []; comments = []

for i in range(950):
    try : 
        # Ïú†Ï†Ä Ïù¥Î¶Ñ
        u = driver.find_elements_by_css_selector('h3.ytd-comment-renderer a span')[i].text
        users.append(u)
        # ÎåìÍ∏Ä
        c = driver.find_elements_by_css_selector('#content-text')[i].text
        comments.append(c)        
    except : 
        pass
```

```python
pd.DataFrame({'user id': users, 'comment': comments})
```

```python
driver.close() 
```

# ÏïÑÏù¥Îîî, ÎåìÍ∏Ä
## xpath

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.youtube.com/watch?v=aN5s9N_pTUs')

time.sleep(3) 

# Ïä§ÌÅ¨Î°§ ÎÇ¥Î¶¨Í∏∞
e = driver.find_element_by_tag_name('body')
for i in range(10): # 30Î≤à
    e.send_keys(Keys.PAGE_DOWN) # ÏïÑÎûò Î≤ÑÌäº
    time.sleep(2) 
```

```python
users = []; comments = []

for i in range(950):
    try : 
        # Ïú†Ï†Ä Ïù¥Î¶Ñ
        u = driver.find_elements_by_xpath('//*[@id="author-text"]/span')[i].text
        users.append(u)
        # ÎåìÍ∏Ä
        c = driver.find_elements_by_xpath('//*[@id="content-text"]')[i].text
        comments.append(c)        
    except : 
        pass
```

```python
pd.DataFrame({'user id': users, 'comment': comments})
```
